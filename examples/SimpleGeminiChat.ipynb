# Simple AloeCraft Gemini Chat

Useful for testing connection to Gemini. This example is single message, single response. No chat history is retained. You can download from [GitHub](https://github.com/aloecraft/AloeGraph) and run locally or run in Google Colab

**Quick Start:** Supply `GEMINI_API_KEY` as specifed in the Setup section and click `Run All`

## 1. Set up Gemini API and key

```python
import sys
from google import genai
GEMINI_API_KEY = None
if 'google.colab' in sys.modules:
  from google.colab import userdata
  GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
  # Download AloeGraph from GitHub if not running locally
  !pip install git+https://github.com/aloecraft/AloeGraph.git
else:
  import os
  GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
  
client = genai.Client(api_key=GEMINI_API_KEY)
```

## 2. Use `NotebookChatDialog` to send and receive messages

```python
from aloegraph.agent.simple_gemini_chat_agent import GeminiChatAgent
from aloegraph.dialog.notebook_chat_dialog import NotebookChatDialog

notebook_dialog = NotebookChatDialog()
GeminiChatAgent.factory(client, notebook_dialog)
notebook_dialog.display()
```
